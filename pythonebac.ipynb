### **2.3. Dados faltantes** 

Colunas que possuem dados faltantes:

```python
df.isna().any()
```

Estatisticas sobre as colunas dos dados faltantes.

```python
def stats_dados_faltantes(df: pd.DataFrame) -> None:

  stats_dados_faltantes = []
  for col in df.columns:
    if df[col].isna().any():
      qtd, _ = df[df[col].isna()].shape
      total, _ = df.shape
      dict_dados_faltantes = {col: {'quantidade': qtd, "porcentagem": round(100 * qtd/total, 2)}}
      stats_dados_faltantes.append(dict_dados_faltantes)

  for stat in stats_dados_faltantes:
    print(stat)
```

```python
stats_dados_faltantes(df=df)
```

```python
stats_dados_faltantes(df=df[df['default'] == 0])
```

```python
stats_dados_faltantes(df=df[df['default'] == 1])
```

Função `lambda` para tratar os dados.

```python
fn = lambda valor: float(valor.replace(".", "").replace(",", "."))

valores_originais = ['12.691,51', '8.256,96', '3.418,56', '3.313,03', '4.716,22']
valores_limpos = list(map(fn, valores_originais))

print(valores_originais)
print(valores_limpos)
```

Aplicando nas colunas de interesse:

```python
df['valor_transacoes_12m'] = df['valor_transacoes_12m'].apply(fn)
df['limite_credito'] = df['limite_credito'].apply(fn)

```

```python
coluna = 'escolaridade'
titulos = ['Escolaridade dos Clientes', 'Escolaridade dos Clientes Adimplentes', 'Escolaridade dos Clientes Inadimplentes']
eixo = 0
max_y = 0
max = df.select_dtypes('object').describe()[coluna]['freq'] * 1.1
figura, eixos = plt.subplots(1, 3, figsize=(20, 5), sharex=True)
for dataframe in [df, df_adimplente, df_inadimplente]:
   
    df_to_plot = dataframe[coluna].value_counts().reset_index()  
    df_to_plot.columns = [coluna, 'frequencia_absoluta']
    df_to_plot.sort_values(by=[coluna], inplace=True)    
    f = sns.barplot(x=coluna, y='frequencia_absoluta', data=df_to_plot, ax=eixos[eixo])
    f.set(title=titulos[eixo], xlabel=coluna.capitalize(), ylabel='Frequência Absoluta')
    f.set_xticklabels(labels=df_to_plot[coluna], rotation=90)
    _, max_y_f = f.get_ylim()
    max_y = max_y_f if max_y_f > max_y else max_y
    f.set(ylim=(0, max_y))
    eixo += 1
   
figura.show()
```

**Conclusão:**

	A análise dos dados revelou que a escolaridade não é um fator relevante para a adimplência dos clientes. No entanto, os atributos numéricos qtd_transacoes_12m e valor_transacoes_12m mostraram uma relação significativa com a adimplência. Clientes com muitas transações (acima de 80) e valores altos (acima de 12 mil) tendem a ser adimplentes. Além disso, foram identificadas combinações específicas de quantidade e valor de transações que indicam tendências de adimplência ou inadimplência.

	A idade também mostrou ser um fator importante, com clientes entre 40 e 50 anos apresentando maior tendência a inadimplência. O atributo iteracao_12m também é relevante, pois quando a quantidade de iterações muda de 2 para 3, há um indicativo de que o cliente possa se tornar inadimplente. Isso é ainda mais evidente em clientes entre 40 e 50 anos que possuem 4 iterações. Em contraste, os atributos categóricos não conseguiram mapear efetivamente o comportamento de adimplência e inadimplência dos clientes.
